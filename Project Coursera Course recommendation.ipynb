```python
tfidf_vectorizer = TfidfVectorizer(stop_words='english')
tfidf_matrix = tfidf_vectorizer.fit_transform(df['Skills'])
```

```python
cosine_sim = linear_kernel(tfidf_matrix, tfidf_matrix)
```

```python
def get_recommendations(course_name, cosine_sim=cosine_sim):
    idx = df.index[df['Course Name'] == course_name].tolist()[0]
    sim_scores = list(enumerate(cosine_sim[idx]))
    sim_scores = sorted(sim_scores, key=lambda x: x[1], reverse=True)
    sim_scores = sim_scores[1:11]  # Get top 10 similar courses (excluding the input course)
    course_indices = [score[0] for score in sim_scores]
    return df.iloc[course_indices]
```

```python
course_name = input('Please enter the course name :-  ')
recommended_courses = get_recommendations(course_name)
print(f"Recommended courses similar to '{course_name}':")
print(recommended_courses[['Course Name']])
```
